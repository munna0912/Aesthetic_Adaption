# 1. Import libraries and Pretrained Model

```python
import tensorflow_hub as hub
import tensorflow as tf
from matplotlib import pyplot as plt
import numpy as np
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
model = hub.load('https://tfhub.dev/google/magenta/arbitrary-image-stylization-v1-256/2')
```

# 2. Preprocess and load Images

```python
def load_image(img_path):
    img = tf.io.read_file(img_path)
    img = tf.image.decode_image(img, channels=3)
    img = tf.image.convert_image_dtype(img, tf.float32)
    img = img[tf.newaxis, :]
    return img
```

```python
input_image = load_image('/content/drive/MyDrive/assignment2/Munna_Ram .jpg')
style_image = load_image('/content/drive/MyDrive/assignment2/Albrecht_Du╠êrer_4.jpg')
```

# 3. Visualize Images

```python
plt.imshow(np.squeeze(style_image))
plt.show()
```

```python
plt.imshow(np.squeeze(input_image))
plt.show()
```

## 4. Stylize Image

```python
stylized_image = model(tf.constant(input_image), tf.constant(style_image))[0]
```

```python
plt.imshow(np.squeeze(stylized_image))
plt.show()
```

```python
image = tf.keras.preprocessing.image.array_to_img(np.squeeze(stylized_image))
image.save("/content/drive/MyDrive/assignment2/stylized_Munna_Ram.jpg")
```
